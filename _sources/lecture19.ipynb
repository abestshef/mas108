# Lecture 19. More examples of motion with constant acceleration 

We will end this series of lectures on motion with constant acceleration with two examples.

````{prf:example} 
A student descending in a lift at constant speed $V$ throws a ball vertically upwards with speed $U$ relative to the lift. Find the time that elapses before the ball
(a) has zero velocity relative to the lift,
(b) returns to the student's hand.

**Solution.** 
```{toggle}
As always, we start with a sketch (Figure 29).

<img src="mas108_student_in_lift.png" alt="drawing" width="400"/>

*Figure 29: Student in a lift.*

(a) At time $t$, let the velocity of the ball be $\mathbf{v}_{B}(t)=W(t) \mathbf{k}$, where $W(t)$ is a scalar function. At time $t=0$,

$$
\begin{align*}
\mathbf{v}_{B}(t) & =\text { velocity of the ball relative to the lift }+ \text { velocity of the lift } \\
& =U \mathbf{k}-V \mathbf{k}=(U-V) \mathbf{k} . \tag{237}
\end{align*}
$$

Since we have constant acceleration, we can use the 'suvat' equations. In particular, we use Equation (211), which says that $\mathbf{v}=\mathbf{u}+\mathbf{a} t$. Here, $\mathbf{u}=\mathbf{v}_{B}(0)$ and $\mathbf{a}=g \mathbf{k}$, so the velocity of the ball at time $t$ is $\mathbf{v}_{B}(t)=\mathbf{v}_{B}(0)-g t \mathbf{k}$. Therefore, combing Equation (237) and $\mathbf{v}_{B}(t)=W(t) \mathbf{k}$, we have

$$
\begin{align*}
& W(t) \mathbf{k}=(U-V) \mathbf{k}-g t \mathbf{k}  \tag{238}\\
\Longrightarrow & W(t)=U-V-g t \tag{239}
\end{align*}
$$

Suppose the ball has zero velocity relative to the lift at time $t=t_{1}$. Then the velocity of the ball is the same as the velocity of the lift, i.e. $\mathbf{v}_{B}\left(t_{1}\right)=-V \mathbf{k}$. Since $\mathbf{v}_{B}(t)=W(t) \mathbf{k}$, we have $W\left(t_{1}\right)=-V$. Then, by Equation (239), $U-V-g t_{1}=-V$. Consequently, $t_{1}=U / g$, i.e. the ball has zero velocity relative to the lift at time $t=U / g$.
(b) At time $t$, let the location of the ball, relative to its starting location, be $\mathbf{x}_{B}(t)$. Here, we use Equation (212) to give

$$
\begin{equation*}
\mathbf{x}_{B}(t)=\mathbf{v}_{B}(0) t-\frac{1}{2} g t^{2} \mathbf{k}=\left[W(0) t-\frac{1}{2} g t^{2}\right] \mathbf{k}=\left[(U-V) t-\frac{1}{2} g t^{2}\right] \mathbf{k} \tag{240}
\end{equation*}
$$

where the final equality comes from using Equation (239) with $t=0$.
Denote the location of the student's hand at time $t$ by $\mathbf{x}_{S}(t)$, so that $\mathbf{x}_{S}(t)=-V t \mathbf{k}$. Suppose the ball returns to the student's hand at a time $t=t_{2}$. Then $\mathbf{x}_{B}\left(t_{2}\right)=\mathbf{x}_{S}\left(t_{2}\right)=$ $-V t_{2} \mathbf{k}$. Therefore, using Equation (240) with $t=t_{2}$,

$$
\begin{align*}
-V t_{2}=\mathbf{x}_{B}\left(t_{2}\right) & =(U-V) t_{2}-\frac{1}{2} g t_{2}^{2}  \tag{241}\\
\Longrightarrow & 0=U t_{2}-\frac{1}{2} g t_{2}^{2}=t_{2}\left(U-\frac{1}{2} g t_{2}\right) . \tag{242}
\end{align*}
$$

Therefore either $t_{2}=0$ (time of projection) or

$$
\begin{align*}
U & -\frac{1}{2} g t_{2}=0  \tag{243}\\
\Longrightarrow t_{2} & =\frac{2 U}{g} \tag{244}
\end{align*}
$$

Thus the ball returns to the student's hand at time $t=2 U / \mathrm{g}$.
```
````
````{prf:example}
A particle is projected from the origin with speed $V$ at an angle $\theta$ above the horizontal. There is no air resistance. Show that
(a) The maximum height of the particle above the point of projection is

$$
H=\frac{V^{2}}{2 g} \sin ^{2}(\theta)
$$

(b) The time at which the particle returns to the same vertical level as the point of projection is

$$
T=\frac{2 V}{g} \sin (\theta)
$$

(c) When the particle returns to the same vertical level as the point of projection its horizontal distance from the point of projection is

$$
R=\frac{V^{2}}{g} \sin (2 \theta)
$$

**Solution.** 
```{toggle}
Recall the following formulae for a projectile from Equations (225), (226), (228), and (229)

$$
\begin{align*}
& x(t)=V t \cos (\theta)  \tag{245}\\
& z(t)=V t \sin (\theta)-\frac{1}{2} g t^{2}  \tag{246}\\
& \dot{x}(t)=V \cos (\theta)  \tag{247}\\
& \dot{z}(t)=V \sin (\theta)-g t \tag{248}
\end{align*}
$$

(a) Let $t=t_{H}$ be the time the projectile reaches maximum height. Then $\dot{z}\left(t_{H}\right)=0$, so by

Equation (248),

$$
\begin{equation*}
t_{H}=\frac{V}{g} \sin (\theta) \tag{249}
\end{equation*}
$$

Therefore the maximum height is

$$
\begin{align*}
H=z\left(t_{H}\right) & =V \sin (\theta) \times \frac{V}{g} \sin (\theta)-\frac{1}{2} g\left[\frac{V}{g} \sin (\theta)\right]^{2} \\
& =\frac{V^{2}}{2 g} \sin ^{2}(\theta) \tag{250}
\end{align*}
$$

(b) The particle returns to the level of projection when $z(T)=0$ so, by Equation (246),

$$
\begin{equation*}
0=V T \sin (\theta)-\frac{1}{2} g T^{2}=T\left(V \sin \theta-\frac{1}{2} g T\right) . \tag{251}
\end{equation*}
$$

The solution $T=0$ is the time of projection. Therefore the time when the particle returns to the level of projection is

$$
\begin{equation*}
T=\frac{2 V}{g} \sin (\theta) \tag{252}
\end{equation*}
$$

(c) Using Equation (245) with $t=T$ and the value of $T$ from Equation 252 gives

$$
\begin{align*}
R=x(T) & =V \cos (\theta) \times \frac{2 V}{g} \sin (\theta)=\frac{V^{2}}{g} \times 2 \sin (\theta) \cos (\theta) \\
& =\frac{V^{2}}{g} \sin (2 \theta) \tag{253}
\end{align*}
$$
```
````

A bus starts from rest and moves along a straight road with constant acceleration $f$ until its speed is $V$. The bus then continues at constant speed $V$. When the bus starts, a car is at a distance $b$ behind the bus and is moving in the same direction with constant speed $u$. Find the distance of the car behind the bus at time $t$ after the bus has started
(a) For $0<t<\frac{V}{f}$;
(b) For $t>\frac{V}{f}$.

Show that the car cannot overtake the bus during the period $0<t<\frac{V}{f}$ unless $u^{2}>2 f b$.
Find the minimum distance between the car and the bus in the case when $u^{2}<2 \mathrm{fb}$ and $u<V$.
